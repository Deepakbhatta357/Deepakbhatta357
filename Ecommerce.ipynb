```python
cust_Seg=df.groupby('Segment')['Sales'].sum().reset_index()
fiig=px.bar(cust_Seg,x='Segment',y='Sales')
fiig.show()
```

```python
sales_profit_Seg=df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
fig=go.Figure()
fig.add_trace(go.Bar(x=sales_profit_Seg['Segment'],
                     y=sales_profit_Seg['Sales'],name='Sales'))

fig.add_trace(go.Bar(x=sales_profit_Seg['Segment'],
                     y=sales_profit_Seg['Profit'],name='Profit'))

fig.show()
```

```python
# Calculate Sales-to-Profit Ratio
sales_profit_Seg['Sales_to_Profit_Ratio'] = sales_profit_Seg['Sales'] / sales_profit_Seg['Profit']

# Display the DataFrame
print(sales_profit_Seg)
```

```python
import plotly.express as px

# Calculate Sales-to-Profit Ratio
sales_profit_Seg['Sales_to_Profit_Ratio'] = sales_profit_Seg['Sales'] / sales_profit_Seg['Profit']

# Create Bar Chart
fig = px.bar(sales_profit_Seg,
             x='Segment',
             y='Sales_to_Profit_Ratio',
             text='Sales_to_Profit_Ratio',
             title="Sales-to-Profit Ratio by Segment",
             labels={'Sales_to_Profit_Ratio': 'Sales / Profit Ratio'},
             color='Segment')

# Update layout for better readability
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(yaxis_title="Sales-to-Profit Ratio")

# Show Plot
fig.show()
```
